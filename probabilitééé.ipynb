```python
print(livre.lower())
```

```python
livre= livre.translate(str.maketrans("","", string.punctuation))
print(livre.lower())
```

```python
liste=[livre.lower().split()]
```

```python
liste
```

```python
import matplotlib.pyplot as plt
listeLe=[]
listeUn=[]
listeEt=[]
listeIl=[]
listeEst=[]
indice=[]
for i in range(len(liste)):
    sommeLe=0
    sommeUn=0
    sommeEt=0
    sommeIl=0
    sommeEst=0
    for j in range (1,i):
        if liste[j]=="Le":
            sommeLe=sommeLe+1
        elif liste[j]=="Un":
            sommeUn=sommeUn+1
        elif liste[j]=="Et":
            sommeEt=sommeEt+1
        elif liste[j]=="Il":
            sommeIl=sommeIl+1
        elif liste[j]=="Est":
            sommeEst=sommeEst+1
    listeLe.append(sommeLe)
    listeUn.append(sommeUn)
    listeEt.append(sommeEt)
    listeIl.append(sommeIl)
    listeEst.append(sommeEst)
    
    indice.append(i)
print(i,sommeIl)
plt.subplot(221)
plt.plot(indice,listeLe)
plt.subplot(222)
plt.plot(indice,listeEt)
        
        
    
```
